## Game recordings

It's really only the four middle columns that are made into strings instead of numbers, this is because of the non numeric character found in all of them. 

2M-2A, 3M-3A and 1M-1A all have the same problem with the '-' character being right in the middle.

Meanwhile, FG% and 1% face the problem of the % symbol appended at the end of our floating point number. Which is an easy fix.

So let's go over how we can fix these issues.

```python
def convert_string_to_numbers(df):
    for index, row in df.iterrows():
        if isinstance(row['2M-2A'], str):
            df.at[index, '2M-2A'] = np.array(row['2M-2A'].split('-'), dtype=int)
        if isinstance(row['3M-3A'], str):
            df.at[index, '3M-3A'] = np.array(row['3M-3A'].split('-'), dtype=int)
        if isinstance(row['1M-1A'], str):
            df.at[index, '1M-1A'] = np.array(row['1M-1A'].split('-'), dtype=int)
        try:
            df.at[index, 'FG%'] = float(row['FG%'].rstrip('%'))
        except:
            df.at[index, 'FG%'] = 0
        try:
            df.at[index, '1%'] = float(row['1%'].rstrip('%'))
        except:
            df.at[index, '1%'] = 0

convert_string_to_numbers(df)
df
```

The function above will make the first three columns that hold two values into an array of values, and it will make the percentage columns into a floating point numbers. As we can see above. The fact we converted our 2M-2A column into an array then to a CSV. The comma which seperates the value of this array is gone. Meaning when loading the CSV, the Python interpreter will not be able to tell that these are actually two different values. So let's edit our function to seperate these columns into two columns, 2M and 2A, rather than making it into an array.

All of this is to say that converting the columns above into an array wont work, unless I do some operations every time I load my CSVs. Therefore the best solution is to make new columns for each value split from the - character.

```python
df = pd.read_csv('../1_Data_Collection/datasets/game_data/atlanta-hawks-minnesota-timberwolves-2023-10-31.csv')
df
```

```python
def remodel(df):
    _2M, _2A, _3M, _3A, _1M, _1A = [], [], [], [], [], []
    for index, row in df.iterrows():
        if '2M-2A' in row and isinstance(row['2M-2A'], str):
            _2M.append(int(row['2M-2A'].split('-')[0]))
            _2A.append(int(row['2M-2A'].split('-')[1]))
        if '3M-3A' in row and isinstance(row['3M-3A'], str):
            _3M.append(int(row['3M-3A'].split('-')[0]))
            _3A.append(int(row['3M-3A'].split('-')[1]))
        if '1M-1A' in row and isinstance(row['1M-1A'], str):
            _1M.append(int(row['1M-1A'].split('-')[0]))
            _1A.append(int(row['1M-1A'].split('-')[1]))
        try:
            if 'FG%' in row:
                df.at[index, 'FG%'] = float(row['FG%'].rstrip('%'))
            else:
                df.at[index, 'FG%'] = 0
        except:
            df.at[index, 'FG%'] = 0
        try:
            if '1%' in row:
                df.at[index, '1%'] = float(row['1%'].rstrip('%'))
            else:
                df.at[index, '1%'] = 0
        except:
            df.at[index, '1%'] = 0
    df.insert(3, '2M', _2M)
    df.insert(4, '2A', _2A)
    df.insert(5, '3M', _3M)
    df.insert(6, '3A', _3A)
    df.insert(7, '1M', _1M)
    df.insert(8, '1A', _1A)        
    

remodel(df)
df.columns
```

As we can see above we get more features and more variables to analyze, but what we do get out of this remodelling function is more interpretable parameters/variables. We're able to make all of our columns into the proper data type, rather than doing String parsing whenever we want perform an operation.

## Patterns

Let's try to find something we can work with. The first thing we can identify is one certainty. The team with the most points will always win.
So I think it's fair to not try to corelate something to the number of points.

First I tried looking at the number of rebounds. In this specific case I found that the number of rebounds cannot accurately predict the outcome, since the timberwolves had four more rebounds than the Hawks, and still lost.

However I will say that it's interesting ATL still has slightly higher mean rebounds.

```python
Atl = df[df['PLAYER'] == 'AtlantaHawks']
Min = df[df['PLAYER'] == 'MinnesotaTimberwolves']

Atl_Players = df[df['Team'] == 'AtlantaHawks']
Min_Players = df[df['Team'] == 'MinnesotaTimberwolves']
# Print the total rebounds for each team
print("ATL Reb total: ", Atl['Reb'].sum(), "\n" "MIN Reb total: ", Min['Reb'].sum())
# Print the mean rebounds for each team
print("ATL Reb mean: ", Atl_Players['Reb'].mean(), "\n" "MIN Reb mean: ", Min_Players['Reb'].mean())
# Print the median rebounds for each team
print("ATL Reb median: ", Atl_Players['Reb'].median(), "\n" "MIN Reb median: ", Min_Players['Reb'].median())
# Print the standard deviation of rebounds for each team
print("ATL Reb std: ", Atl_Players['Reb'].std(), "\n" "MIN Reb std: ", Min_Players['Reb'].std())
# Print the variance of rebounds for each team
print("ATL Reb variance: ", Atl_Players['Reb'].var(), "\n" "MIN Reb variance: ", Min_Players['Reb'].var())
# Print the range of rebounds for each team
print("ATL Reb range: ", Atl_Players['Reb'].max() - Atl_Players['Reb'].min(), "\n" "MIN Reb range: ", Min_Players['Reb'].max() - Min_Players['Reb'].min())
# Print IQR of rebounds for each team
print("ATL Reb IQR: ", Atl_Players['Reb'].quantile(0.75) - Atl_Players['Reb'].quantile(0.25), "\n" "MIN Reb IQR: ", Min_Players['Reb'].quantile(0.75) - Min_Players['Reb'].quantile(0.25))
# Print the skewness of rebounds for each team
print("ATL Reb skewness: ", Atl_Players['Reb'].skew(), "\n" "MIN Reb skewness: ", Min_Players['Reb'].skew())
# Print the kurtosis of rebounds for each team
print("ATL Reb kurtosis: ", Atl_Players['Reb'].kurtosis(), "\n" "MIN Reb kurtosis: ", Min_Players['Reb'].kurtosis())
# Print the mode of rebounds for each team
print("ATL Reb mode: ", Atl_Players['Reb'].mode(), "\n" "MIN Reb mode: ", Min_Players['Reb'].mode())
```

So so finding the dispersion and location of each variable wasn't as bad an idea as I thought. I ended up realizing that more players on the ATL team contributed to rebounds. Hence the higher mean, smaller variance, standard deviation and range. 

The most frequent value will typically be the lowest number, this is because most of the team is usually not gathered by the net to try to get the ball.
